```python
%matplotlib inline
import numpy as np
from matplotlib import rc
rc('font', **{'family': 'serif', 'serif': ['Computer Modern'], 'size': 15})
rc('text', usetex=True)
import matplotlib.pyplot as plt
import revreaddy.sim as sim
import time
import revreaddy.utils as utils
import scipy.integrate as si
simulation1 = sim.pySimulation()
simulation2 = sim.pySimulation()
# abbreviation for plotting functions in certain range
def plot(f,x0,xn,dx, **kw):
    xRange = np.arange(x0,xn,dx)
    yRange = np.array(map(f,xRange))
    plt.plot(xRange,yRange, **kw)
```

```python
radius = 1.
diffConst = 1.
forceType = 2 # soft particles
simulation1.new_Type("custom-soft", radius, diffConst, 1., forceType)
simulation1.showTypes()
simulation2.new_Type("custom-soft", radius, diffConst, 1., forceType)
simulation2.showTypes()
```

```python
simulation1.boxsize, simulation2.boxsize = 20., 20.
simulation1.isPeriodic, simulation2.isPeriodic = True, True
simulation1.temperature, simulation2.temperature = 1., 1.
simulation1.repulsionStrength, simulation2.repulsionStrength = 1., 1.
```

```python
simulation1.new_Wall([0.,0.,-1.], [0.,0., 1.], 1., [3])
simulation1.new_Wall([0.,0., 1.], [0.,0.,-1.], 1., [3])
simulation2.new_Wall([0.,0.,-1.], [0.,0., 1.], 1., [3])
simulation2.new_Wall([0.,0., 1.], [0.,0.,-1.], 1., [3])

L = simulation1.boxsize / 2.
for x in np.arange(-L,L,2.):
    for y in np.arange(-L,L,2.):
        position = [x, y, 0.]
        simulation1.addParticle(position, 3)
        simulation2.addParticle(position, 3)
simulation1.getParticleNumber()
simulation2.getParticleNumber()
```

```python
simulation1.isReversible = False
trajname = "data/traj_" + time.strftime("%Y_%m_%d-%H_%M_%S") + ".xyz"
energyname = "data/energy_" + time.strftime("%Y_%m_%d-%H_%M_%S") + ".dat"
probname = "data/prob_" + time.strftime("%Y_%m_%d-%H_%M_%S") + ".dat"
#simulation1.new_Trajectory(trajname)
simulation1.new_Energy(1, energyname)
simulation1.new_ProbabilityDensity(probname, 3, np.arange(-4.,4.,0.1), 2)
simulation1.maxTime = 20000
simulation1.timestep = 1e-2
simulation1.run()

simulation1.writeAllObservablesToFile()
simulation1.deleteAllObservables()
```

```python
energy = np.loadtxt(energyname, skiprows=1)
energies = [np.transpose(energy)]
```

Reversible
----------

```python
simulation2.isReversible = False
trajname = "data/traj_" + time.strftime("%Y_%m_%d-%H_%M_%S") + ".xyz"
energyname = "data/energy_" + time.strftime("%Y_%m_%d-%H_%M_%S") + ".dat"
probname = "data/prob_" + time.strftime("%Y_%m_%d-%H_%M_%S") + ".dat"
#simulation.new_Trajectory(trajname)
simulation2.new_Energy(1, energyname)
simulation2.new_ProbabilityDensity(probname, 3, np.arange(-4.,4.,0.1), 2)
simulation2.maxTime = 20000
simulation2.timestep = 1e-2
simulation2.run()

simulation2.writeAllObservablesToFile()
simulation2.deleteAllObservables()
```

```python
energy = np.loadtxt(energyname, skiprows=1)
energies += [np.transpose(energy)]
```

```python
prob = np.loadtxt(probname)
probs += [prob]
```

```python
#normalize probs to 1.
for i in range(len(probs)):
    norm = si.simps(probs[i][1], probs[i][0])
    probs[i][1] /= norm
```

```python
energyFig, energyAxes = plt.subplots()
energyAxes.plot(energies[0][0], energies[0][1], color="red", label="irreversible")
energyAxes.plot(energies[1][0], energies[1][1], color="green", label="reversible")
energyAxes.set_xlabel("time")
energyAxes.set_ylabel("total energy")
energyAxes.grid(b=1)
energyAxes.legend(loc="lower right")
energyAxes.set_xlim(0.,3.)
```

```python
probFig, probAxes = plt.subplots()
probAxes.plot(probs[0][0], probs[0][1], color="red", label="irreversible")
probAxes.plot(probs[1][0], probs[1][1], color="green", label="reversible")
probAxes.set_xlabel("z")
probAxes.set_ylabel("probability density")
probAxes.grid(b=1)

def membranePot(z):
    if (z < -1.):
        return (z + 1.)**2
    if (z >  1.):
        return (z - 1.)**2
    else:
        return 0.
#def membranePot(z): return z**2

f = lambda x: np.exp(-membranePot(x)) / ( np.sqrt(np.pi) + 2.)
plot(f, -4., 4., 0.1, label="equilibrium distribution")
#probAxes.legend()
```
