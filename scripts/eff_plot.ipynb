```python
%matplotlib inline
import numpy as np
from matplotlib import rc
rc('font', **{'family': 'serif', 'serif': ['Computer Modern'], 'size': 15})
rc('text', usetex=True)
import matplotlib.pyplot as plt
import pylab
pylab.rcParams['figure.figsize'] = (8.0, 8.0)
```

```python
eff_lj    = np.loadtxt("data/eff/eff_lj.dat")
taus_lj   = eff_lj[0]
effs_lj   = eff_lj[1]
accs_lj   = eff_lj[2]
eff_soft  = np.loadtxt("data/eff/eff_soft.dat")
taus_soft = eff_soft[0]
effs_soft = eff_soft[1]
accs_soft = eff_soft[2]
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1)
etaAxes = axes[0]
etaAxes.grid(b=1)
etaAxes.plot(taus_lj, effs_lj, "bo",label="Lennard-Jones")
etaAxes.plot(taus_soft, effs_soft, "ro", label="Soft")
etaAxes.set_xscale("log")
etaAxes.set_yscale("log")
etaAxes.legend(loc="upper left")
etaAxes.set_ylabel(r"Efficiency $\eta$")
ylim = list(etaAxes.get_ylim())
ylim[0] = 1e-7
etaAxes.set_ylim(ylim)
accAxes = axes[1]
accAxes.grid(b=1)
accAxes.plot(taus_lj, accs_lj, "bo")
accAxes.plot(taus_soft, accs_soft, "ro")
accAxes.set_xscale("log")
accAxes.set_xlabel(r"Timestep $\tau$")
accAxes.set_ylabel(r"Acceptance $a$")
fig.savefig("data/eff/result.pdf", dpi=300)
```
